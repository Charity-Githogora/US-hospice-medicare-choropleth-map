# **State-Level Data Preparation for Hospice Choropleth Map**

This notebook prepares state-level aggregated hospice metrics from the CMS Medicare Hospice dataset.
The output is used to create a choropleth map in Power BI.

```python
import pandas as pd
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_excel("US Hospice Medicare Data.xlsx")
```

```python
df.columns = (
    df.columns
      .str.lower()
      .str.replace(" ", "_")
      .str.replace("/", "_")
)
```

```python
df['state'].nunique(), df['state'].sort_values().unique()
```

```python
state_summary = (
    df.groupby("state")
      .agg(
          total_providers=("provider_id", "nunique"),
          total_beneficiaries=("hospice_beneficiaries", "sum"),
          total_medicare_payment=("total_medicare_payment_amount", "sum")
      )
      .reset_index()
)
```

```python
state_summary.to_csv(
    "state_summary.csv",
    index=False
)
```
